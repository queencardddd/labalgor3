```python
def linear_search(arr, target):
    # Проходим по списку по очереди
    for i in range(len(arr)):

        # Если нашли нужное число
        if arr[i] == target:
            return i  # возвращаем индекс

    # Если число не нашли
    return -1


# Проверка
print(linear_search([10, 50, 30, 70, 80, 20], 30))
```
